```python
import matplotlib.pyplot as plt
import pandas as pd

csv_path = "training_2024.01.22.14.44.59/lightning_logs/version_0/metrics.csv"


df = pd.read_csv(csv_path)

# Plot train_loss vs epoch
plt.plot(df['epoch'], df['train_loss'], label='val_loss_epoch')
plt.xlabel('Epoch')
plt.ylabel('train_loss_epoch')
plt.yscale("log")
plt.title('Train Loss vs Epoch (QCD_HT_2000toInf)')
plt.legend()
plt.show()
```
